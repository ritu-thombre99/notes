```python
import qiskit
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, BasicAer, execute
import numpy as np
```

```python
backend = BasicAer.get_backend("statevector_simulator")
```

```python
qc1 = QuantumCircuit(5)
qc1.h(0)
for i in range(1,5):
    qc1.cnot(0,i)
qc1.draw(output='mpl')
```

```python
execute(qc1,backend).result().get_statevector()
```

```python
from qiskit.extensions import UnitaryGate
matrix = [[(1-1j)/np.sqrt(2), 0, 0, 0],
          [0, (1+1j)/np.sqrt(2), 0, 0],
          [0, 0, (1+1j)/np.sqrt(2), 0],
          [0, 0, 0, (1-1j)/np.sqrt(2)]]
print(matrix)
gate = UnitaryGate(matrix)
```

```python
qc2 = QuantumCircuit(5)
for i in range(5):
    qc2.h(i)


for i in range(1,5):
    qc2.rz(np.pi/2,i)
    qc2.append(gate,[0,i])
qc2.barrier()
for i in range(1,5):
    qc2.h(i)
qc2.x(0)
qc2.draw(output='mpl')
```

```python
execute(qc2,backend).result().get_statevector()
```
